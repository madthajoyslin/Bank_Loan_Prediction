### Outlier using Statistical approach

### feature is skewed compute IQR

```python
# Select numerical columns excluding 'loan_status'
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
if 'loan_status' in numerical_cols:
    numerical_cols.remove('loan_status')  # Exclude loan_status
plt.figure(figsize=(15, len(numerical_cols) * 5))  # Set figure size

# Finding outliers using IQR
for i, col in enumerate(numerical_cols, 1):
    Q1 = df[col].quantile(0.25)  # First quartile
    Q3 = df[col].quantile(0.75)  # Third quartile
    IQR = Q3 - Q1  # Interquartile range

    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]  # Extract outlier values

    print(f"\n{col}: {len(outliers)} outliers detected")
    print(outliers)  # Display outlier values with their indices

    # Create a boxplot to visualize outliers
    plt.subplot(len(numerical_cols), 1, i)
    sns.boxplot(x=df[col], color='skyblue')
    
    # Highlighting outliers as red dots
    sns.stripplot(x=outliers, color='red', size=5, jitter=True)
    
    plt.title(f"Boxplot of {col} (Outlier Detection)")

plt.tight_layout()
plt.show()
```

## Train test split 
### SMOTE can be applied after Train Test split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer
from imblearn.over_sampling import SMOTE
from collections import Counter
```

### 1. Separate Features (X) and Target (y)

```python
# Using df_encoded as it contains the one-hot encoded features and the mapped 'previous_loan_defaults_on_file'
X = df_encoded.drop('loan_status', axis=1)
y = df_encoded['loan_status']
```

### Train test split 

```python
# Store original feature names for later use (feature importance)
original_feature_names = X.columns.tolist()

# 2. Split Data into Training and Testing Sets
# It's crucial to split *before* applying SMOTE to avoid data leakage
# We use stratify=y to maintain the original class proportion in the test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

```

### Applying chi square for feature selection
#### Have very low p-values (typically < 0.05)

```python
from sklearn.feature_selection import SelectKBest, chi2
# Step 1: Select only categorical (non-continuous) columns
# You can identify categorical features by their dtype or if they're one-hot encoded
categorical_cols = X.select_dtypes(include=['uint8', 'bool']).columns

# Filter X to only keep categorical columns
X_categorical = X[categorical_cols]

# Step 2: Apply Chi-Square Feature Selection
chi2_selector = SelectKBest(score_func=chi2, k='all')
X_chi2 = chi2_selector.fit_transform(X_categorical, y)

# Step 3: Get feature scores and names
chi2_scores = chi2_selector.scores_
chi2_pvalues = chi2_selector.pvalues_
chi2_features = X_categorical.columns[chi2_selector.get_support()]

# Step 4: View feature scores
chi2_results_df = pd.DataFrame({
    'feature': X_categorical.columns,
    'chi2_score': chi2_scores,
    'p_value': chi2_pvalues
}).sort_values(by='chi2_score', ascending=False)

print("\n--- Top Chi-Square Scores for Categorical Features ---")
print(chi2_results_df.head(30))
```

### Excluded features with p-value >= 0.05

```python
# Filter features with p-value less than 0.05
important_features = chi2_results_df[chi2_results_df['p_value'] >= 0.05]
print(important_features[['feature', 'chi2_score', 'p_value']])
```

### Anova F-test for the numerical values

```python
from sklearn.feature_selection import SelectKBest, f_classif

# Step 1: Select only numerical (continuous) columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
X_numerical = X[numerical_cols]

# Step 2: Apply ANOVA F-test Feature Selection
anova_selector = SelectKBest(score_func=f_classif, k='all')
X_anova = anova_selector.fit_transform(X_numerical, y)

# Step 3: Get F-scores and p-values
anova_scores = anova_selector.scores_
anova_pvalues = anova_selector.pvalues_
anova_features = X_numerical.columns[anova_selector.get_support()]

# Step 4: View results
anova_results_df = pd.DataFrame({
    'feature': X_numerical.columns,
    'f_score': anova_scores,
    'p_value': anova_pvalues
}).sort_values(by='f_score', ascending=False)

print("\n--- Top ANOVA F-Scores for Numerical Features ---")
print(anova_results_df.head(30))
```

### excluded features with anova F-test

```python
# Get all numerical feature names
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Get selected features using the ANOVA selector
selected_numerical_features = X[numerical_cols].columns[anova_selector.get_support()]

# Find features that were NOT selected
excluded_numerical_features = set(numerical_cols) - set(selected_numerical_features)

print("\n Features NOT selected by ANOVA F-test:")
print(list(excluded_numerical_features))
```

### SMOTE

```python
print("--- Class Distribution Before SMOTE ---")
print("Original training set shape %s" % Counter(y_train))
print("Original test set shape %s" % Counter(y_test))
print("Original X_train shape:", X_train.shape)
print("Original X_test shape:", X_test.shape)

# 3. Apply SMOTE to the Training Data Only
print("\nApplying SMOTE to the training data...")
smote = SMOTE(random_state=42) # random_state for reproducibility
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("\n--- Class Distribution After SMOTE (Training Set) ---")
print("Resampled training set shape %s" % Counter(y_train_resampled))
print("Resampled X_train shape:", X_train_resampled.shape)
# The test set remains unchanged
print("Test set shape remains %s" % Counter(y_test))
print("Test X shape remains:", X_test.shape)
```

### Apply Feature Scaling (StandardScaler)

```python
# Fit the scaler ONLY on the resampled training data
# Transform BOTH the resampled training data and the original test data
print("\nApplying StandardScaler...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_resampled)
X_test_scaled = scaler.transform(X_test) # Use transform, not fit_transform on test data

print("Shape after scaling (train):", X_train_scaled.shape)
print("Shape after scaling (test):", X_test_scaled.shape)

# Convert scaled arrays back to DataFrames (optional, but useful for viewing)
# Use original_feature_names here
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=original_feature_names)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=original_feature_names)
## the actual data is stored in these variables after scaling
```

### Saving scaled data into pickle file. Then load where unseen data is there

```python
import joblib
scaler_filename = 'standard_scaler.pkl'
joblib.dump(scaler, scaler_filename)
print(f"Scaler saved to '{scaler_filename}'")
```

```python
# 5. Feature Importance using RandomForestClassifier
# Train a model on the scaled, resampled data to get feature importances
print("\nCalculating Feature Importance using RandomForestClassifier...")
rf_model = RandomForestClassifier(n_estimators=150, random_state=42, n_jobs=-1, class_weight='balanced') # Added class_weight just in case, though SMOTE handles balance
rf_model.fit(X_train_scaled, y_train_resampled)
#You're training the Random Forest model:

# Get feature importances
importances = rf_model.feature_importances_
feature_names = original_feature_names # Use the names stored earlier

# Create a DataFrame for visualization
feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})

# Sort features by importance
feature_importance_df = feature_importance_df.sort_values('importance', ascending=False).reset_index(drop=True)

print("\n--- Top 10 Feature Importances ---")
print(feature_importance_df.head(10))

# Plot feature importances
plt.figure(figsize=(10, 8))
sns.barplot(x='importance', y='feature', data=feature_importance_df)
plt.title('Feature Importance from RandomForestClassifier (after SMOTE & Scaling)')
plt.xlabel('Importance Score')
plt.ylabel('Features')
plt.tight_layout()
plt.show()
```

### Model training

```python
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
from collections import Counter
```

#### --- Training RandomForest ---

```python
stratified_kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
results = {}
# --- Training RandomForest ---
print("\n--- Training RandomForest ---")
rf_params = {'n_estimators': [150,200], 'max_depth': [20,25], 'min_samples_split': [2,5]}
rf_grid_search = GridSearchCV(RandomForestClassifier(random_state=42, class_weight='balanced'),
                              param_grid=rf_params,
                              cv=stratified_kfold,
                              scoring='accuracy',
                              n_jobs=-1,
                              verbose=1)
print("Running GridSearchCV for RandomForest...")
rf_grid_search.fit(X_train_scaled, y_train_resampled)
print(f"Best Parameters for RandomForest: {rf_grid_search.best_params_}")
print(f"Best Cross-Validation Score (Accuracy) for RandomForest: {rf_grid_search.best_score_:.4f}")
best_rf_model = rf_grid_search.best_estimator_
results['RandomForest'] = rf_grid_search

# Make predictions on the training set
y_train_pred_rf = best_rf_model.predict(X_train_scaled)
train_accuracy_rf = accuracy_score(y_train_resampled, y_train_pred_rf)
print(f"Training Accuracy for RandomForest: {train_accuracy_rf:.4f}")

# Make predictions on the test set
y_test_pred_rf = best_rf_model.predict(X_test_scaled)
test_accuracy_rf = accuracy_score(y_test, y_test_pred_rf)
print(f"Testing Accuracy for RandomForest: {test_accuracy_rf:.4f}")
```

#### --- Training XGBoost --

```python
# --- Training XGBoost ---
print("\n--- Training XGBoost ---")
xgb_params = {'n_estimators': [150,200], 'max_depth': [7,10], 'learning_rate': [0.1,0.05]}
xgb_grid_search = GridSearchCV(XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'),
                               param_grid=xgb_params,
                               cv=stratified_kfold,
                               scoring='accuracy',
                               n_jobs=-1,
                               verbose=2)
print("Running GridSearchCV for XGBoost...")
xgb_grid_search.fit(X_train_scaled, y_train_resampled)
print(f"Best Parameters for XGBoost: {xgb_grid_search.best_params_}")
print(f"Best Cross-Validation Score (Accuracy) for XGBoost: {xgb_grid_search.best_score_:.4f}")
best_xgb_model = xgb_grid_search.best_estimator_
results['XGBoost'] = xgb_grid_search

# Make predictions on the training data
y_train_pred_xgb = best_xgb_model.predict(X_train_scaled)
train_accuracy_xgb = accuracy_score(y_train_resampled, y_train_pred_xgb) # Use y_train_resampled if that's what you trained on
print(f"Training Accuracy for XGBoost: {train_accuracy_xgb:.4f}")

# Make predictions on the testing data
y_test_pred_xgb = best_xgb_model.predict(X_test_scaled)
test_accuracy_xgb = accuracy_score(y_test, y_test_pred_xgb)
print(f"Testing Accuracy for XGBoost: {test_accuracy_xgb:.4f}")
```
